*   http://www-personal.umich.edu/~mejn/netdata/　からGML形式のネットワークデータ(lesmis.gmlなど)をダウンロード
*   以下のコードを実行(右上の「接続」を押してから、左下の黒丸の中に三角が描いてあるボタンを押す)
*   黒丸のところでぐるぐる回り出したら、「参照...」のボタンを押して、先ほどのデータを選択
*   選択したネットワークが可視化されて終了

```
import networkx as nx
from google.colab import files
uploaded = files.upload()
for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))
  G = nx.readwrite.gml.read_gml(fn)
  nx.draw_spring(G, node_size=400, node_color='red', with_labels=True, font_weight='bold')
```
